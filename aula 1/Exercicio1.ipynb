# 1
Gerar listas com valores aleatórios, simulando alturas e pesos e salvar as informações em um arquivo .txt;


```
import random
```

```
pesos, alturas = [], []
```

```
pesos = [round(random.uniform(50, 90), 2) for _ in range(1000)]
alturas = [round(random.uniform(1.5, 2.0), 2) for _ in range(1000)]
```

```
print(pesos[:10])
print(alturas[:10])
```

```
with open("file.txt", "w") as file:
  file.write("Pesos;Alturas\n")
  for peso, altura in zip(pesos, alturas):
    #file.write("%s;%s\n" % (str(peso), str(altura)))
    #file.write(str(peso) + ";" + str(altura) + "\n")
    file.write("{};{}\n".format(peso, altura))
```

# 2
Ler arquivo .txt gerado e exibir as listas em forma de tabela;


```
file = open("file.txt", "r")
lines = file.readlines()
file.close()
```

```
for line in lines[:10]:
  p, a = line.split(";")
  print("{}\t|\t{}".format(p, a))
```

# 3
Calcular o IMC e a classificação de saúde de cada usuário registrado;


```
def classific(imc):
  if imc < 18.5:
    return 1
  elif imc >= 18.5 and imc <= 24.9:
    return 2
  elif imc >= 25.0 and imc <= 29.9:
    return 3
  elif imc >= 30.0 and imc <= 34.9:
    return 4
  elif imc >= 35.0 and imc <= 39.9:
    return 5
  elif imc >= 40.0:
    return 6
```

```
imcs, classificacoes = [], []
for line in lines[1:]:
  p, a = line.split(";")
  imc = float(p) / (float(a)**2)
  imcs.append(imc)
  classificacoes.append(classific(imc))
```

# 4
Exibir novamente as listas em formato de tabela com uma coluna extra exibindo o IMC;


```
for line, imc, classificacao in zip(lines[1:], imcs, classificacoes):
  p, a = line.split(";")
  print("{}\t|\t{}\t|\t{}\t|\t{}".format(p, a.strip(), imc, classificacao))
```
